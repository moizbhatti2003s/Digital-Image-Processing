```python
import cv2
from matplotlib import pyplot as plt

# Function to binarize the image
def binarize_image(gray_image):
    # Apply thresholding with a threshold of 127
    ret, binarized_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
    return binarized_image

# Load the image
image = cv2.imread('image.png')

# Find size and number of channels
image_shape = image.shape
height, width = image_shape[:2]
channels = image_shape[2] if len(image_shape) == 3 else 1

# Display the image information
print(f"Image Size: {height}x{width}")
print(f"Number of Channels: {channels}")

# If the image is a three-channel (color) image
if channels == 3:
    print("This is a three-channel image (color).")

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Binarize the grayscale image
    binarized_image = binarize_image(gray_image)

    # Display grayscale and binarized images
    plt.figure(figsize=(10, 5))

    # Display grayscale image
    plt.subplot(1, 2, 1)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Grayscale Image')
    plt.axis('off')

    # Display binarized image
    plt.subplot(1, 2, 2)
    plt.imshow(binarized_image, cmap='gray')
    plt.title('Binarized Image')
    plt.axis('off')

    plt.show()

# If the image is a one-channel (grayscale) image
elif channels == 1:
    print("This is a one-channel image (grayscale).")

    # Display the image
    plt.imshow(image, cmap='gray')
    plt.title('Grayscale Image')
    plt.axis('off')
    plt.show()

else:
    print("Unknown number of channels.")
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image in grayscale
image = cv2.imread('image.png', cv2.IMREAD_GRAYSCALE)

# Define the threshold value
threshold_value = 127

# Get the dimensions of the image
height, width = image.shape

# Create an empty array for the binarized image
binarized_image = np.zeros((height, width), dtype=np.uint8)

# Loop through each pixel in the image
for i in range(height):
    for j in range(width):
        # Compare each pixel with the threshold value
        if image[i, j] > threshold_value:
            binarized_image[i, j] = 255  # Set to white
        else:
            binarized_image[i, j] = 0  # Set to black

# Display the original and binarized images side by side
plt.figure(figsize=(10, 5))

# Display original grayscale image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

# Display manually binarized image
plt.subplot(1, 2, 2)
plt.imshow(binarized_image, cmap='gray')
plt.title('Binarized Image (Manual)')
plt.axis('off')

plt.show()
```

```python
import cv2

# Open the webcam (0 is the default webcam)
cap = cv2.VideoCapture(0)

# Check if the webcam is opened successfully
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # If the frame is read correctly, ret will be True
    if not ret:
        print("Error: Can't receive frame (stream end?). Exiting ...")
        break

    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Display the grayscale frame
    cv2.imshow('Grayscale Webcam Video', gray_frame)

    # Press 'q' to exit the video window
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close windows
cap.release()
cv2.destroyAllWindows()
```
